# Network

```python
df.mp_plot.network(
      source_col="movieId",
      target_col="userId",
      source_attrs=["title",  "tag"],
      target_attrs=['age', 'gender', 'zip-code', 'occupation'],  
      edge_attrs=["rating"],
      source_color = '#5F1A1F',
      target_color = '#C8B519'
)
```

```python
G = nx.from_pandas_edgelist(df, source='movieId', target='userId', edge_attr=True)
```

```python
print('Number of nodes:', G.number_of_nodes())
print('Number of edges:', G.number_of_edges())
```

### Characteristics

```python
density = nx.density(G)
transitivity = nx.transitivity(G)
avg_clustering = nx.average_clustering(G)

print("Density: %s" % density)
print("Transitivity: %s" % transitivity)
print("Avg. Clustering: %s" % avg_clustering)
```

```python
eigenvector_centrality = nx.eigenvector_centrality(G, max_iter=600)
print("Eigenvalue centrality: %s" % eigenvector_centrality)
```

```python
degree_centrality   = nx.degree_centrality(G)
print("Degree centrality: %s" % degree_centrality)
```

```python
betweeness_centrality = nx.betweenness_centrality(G)
print("Betweeness centrality: %s" % betweeness_centrality)
```

### Communities

```python
modularity = nx_comm.modularity(G, nx_comm.label_propagation_communities(G))
modularity
```

```python
partition= community_louvain.best_partition(G)
```

```python
partition_df = pd.DataFrame(partition.values(), index=partition.keys(), columns=['Community'])
partition_df = partition_df.reset_index(level=0)
partition_df = partition_df.rename(columns={"index": "movieId"})
partition_df.head(20)
```

```python
print('Number of communities: ', partition_df['Community'].nunique())
```

```python
partition_df.groupby('Community')["movieId"].count()
```

```python
df['gender'] = [0 if x=='F' else 1 for x in df['gender']]
```

```python
df['zip-code'] = [x.replace("-","") for x in df['zip-code']]
df['zip-code'] = df['zip-code'].astype(int)
```

```python
def clean_title(a,b):
  a = a.replace(b, "")
  return a

df['Year'] = [x.split(" ")[-1] for x in df['title']]
df['title'] = df.apply(lambda x: clean_title(a = x['title'], b = x['Year']), axis=1)
df['Year'] = [x.replace(")", "").replace("(","") for x in df['Year']]
df['Year'] = df['Year'].astype(int)
```

```python
df['genre'] = df['genre'].astype(str)
df['genre'] = [x.split("|") for x in df['genre']]
```

```python
from sklearn.preprocessing import MultiLabelBinarizer

genres = df['genre']
mlb = MultiLabelBinarizer()
df = df.join(df['genre'].str.join('|').str.get_dummies())
```

```python
df.head()
```

```python
X = df.drop(['rating', 'genre', 'title', 'tag'], axis = 1)
y = df['rating'] 
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 0)
  
from sklearn.naive_bayes import GaussianNB
gnb = GaussianNB().fit(X_train, y_train)
gnb_predictions = gnb.predict(X_test)

accuracy = gnb.score(X_test, y_test)
accuracy
```

```python
cm = confusion_matrix(y_test, gnb_predictions)
cm
```
